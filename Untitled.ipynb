
1. Dropping null values: This approach involves removing rows or columns that have missing values. This can be done using the `dropna()` function in pandas. However, this approach should be used with caution as it may lead to loss of valuable information if there are a significant number of missing values.

2. Imputation: Imputation involves replacing missing values with estimated values. This can be done using various techniques such as mean imputation (replacing missing values with the mean of the column), median imputation (replacing missing values with the median of the column), or mode imputation (replacing missing values with the mode of the column). Imputation helps to retain the existing data and can be useful when the missing values are not random.

3. Splitting into train and test sets: As mentioned earlier, you can split the dataset into two sets based on the presence of missing values. This allows you to handle missing values separately for training and testing purposes.

4. Advanced imputation methods: There are more advanced imputation techniques available, such as using machine learning algorithms to predict missing values based on other features in the dataset. These methods can be more accurate but also more complex to implement.

The choice of approach depends on various factors, including the nature of the data, the amount and pattern of missing values, and the specific requirements of your analysis or modeling task. It's important to carefully consider the implications of each approach and choose the one that best suits your needs.


##### New approch: Splitting the null & non-null values of Purchase col into ['train' & 'test'] & Feature scaling it
 - Aim: Try a diffrent approch for handling null values

```python
'''
df_test: This dataframe contains rows where the 'Purchase' column has null values. It can be considered as a test dataset for further analysis or evaluation.

df_train: This dataframe contains rows where the 'Purchase' column does not have null values. It can be used as a training dataset for modeling or further analysis.
'''
df_test = df2[df2['Purchase'].isnull()]
df_train = df2[~df2["Purchase"].isnull()]
```

```python
# In this case, since the df_test dataset contains instances with non-null values, it is not needed for the train-test split.

# The train-test split should be performed on the df_train dataset, and then the feature variables should be standardized. Here's an example of how you can do it:
```

```python
from sklearn.model_selection import train_test_split
# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df_train.drop('Purchase', axis=1), df_train['Purchase'], test_size=0.2, random_state=42)
```

#### Feauture Scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler #imporitng StandardScaler
# Applying standard scaling to the feature variables
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# using scandard scaler , it will scale the X_test_scaled = scaler.transform(X_test) inside X_train_scaled = scaler.fit_transform(X_train? mean of 0 scale of 1
```

```python
# Print summary statistics before scaling
print("Original Variable Statistics:")
print("X_train - Mean:", X_train.mean(axis=0))
print("X_train - Standard Deviation:", X_train.std(axis=0))
print("X_test - Mean:", X_test.mean(axis=0))
print("X_test - Standard Deviation:", X_test.std(axis=0))

# Print summary statistics after scaling
print("\nScaled Variable Statistics:")
print("X_train_scaled - Mean:", X_train_scaled.mean(axis=0))
print("X_train_scaled - Standard Deviation:", X_train_scaled.std(axis=0))
print("X_test_scaled - Mean:", X_test_scaled.mean(axis=0))
print("X_test_scaled - Standard Deviation:", X_test_scaled.std(axis=0))
```

```python
import matplotlib.pyplot as plt

# Plot histograms before scaling
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.hist(X_train.values[:, 0], bins=20, edgecolor='black')
plt.title('X_train - Variable 1 (Before Scaling)')

# Plot histograms after scaling
plt.subplot(2, 2, 2)
plt.hist(X_train_scaled[:, 0], bins=20, edgecolor='black')
plt.title('X_train_scaled - Variable 1 (After Scaling)')

plt.tight_layout()
plt.show()
```

### Model Selection

- duplicate the scaled variables X_train_scaled and X_test_scaled and apply different regression models   to predict the purchase amount:

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
# Initialize regression models
linear_reg = LinearRegression()
ridge_reg = Ridge()
lasso_reg = Lasso()
decision_tree_reg = DecisionTreeRegressor()
random_forest_reg = RandomForestRegressor()
gradient_boosting_reg = GradientBoostingRegressor()
svr_reg = SVR()
```

```python
# Duplicate the scaled variables
X_train_scaled_dup = X_train_scaled.copy()
X_test_scaled_dup = X_test_scaled.copy()
```

```python
# Fit the regression models on the training data
linear_reg.fit(X_train_scaled_dup, y_train)
ridge_reg.fit(X_train_scaled_dup, y_train)
lasso_reg.fit(X_train_scaled_dup, y_train)
decision_tree_reg.fit(X_train_scaled_dup, y_train)
random_forest_reg.fit(X_train_scaled_dup, y_train)
gradient_boosting_reg.fit(X_train_scaled_dup, y_train)
svr_reg.fit(X_train_scaled_dup, y_train)
```

```python
# Predict purchase amount using the trained models
linear_reg_preds = linear_reg.predict(X_test_scaled_dup)
ridge_reg_preds = ridge_reg.predict(X_test_scaled_dup)
lasso_reg_preds = lasso_reg.predict(X_test_scaled_dup)
decision_tree_reg_preds = decision_tree_reg.predict(X_test_scaled_dup)
random_forest_reg_preds = random_forest_reg.predict(X_test_scaled_dup)
gradient_boosting_reg_preds = gradient_boosting_reg.predict(X_test_scaled_dup)
svr_reg_preds = svr_reg.predict(X_test_scaled_dup)
```

```python
print("Linear Regression Predictions:", linear_reg_preds)
print("Ridge Regression Predictions:", ridge_reg_preds)
print("Lasso Regression Predictions:", lasso_reg_preds)
print("Decision Tree Regression Predictions:", decision_tree_reg_preds)
print("Random Forest Regression Predictions:", random_forest_reg_preds)
print("Gradient Boosting Regression Predictions:", gradient_boosting_reg_preds)
print("SVR Regression Predictions:", svr_reg_preds)
```
